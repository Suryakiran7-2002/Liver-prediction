```python
# our gender column has values entered as "male" and "female", hence lets create dummy variable in overcome this problem
```

```python
dummies = pd.get_dummies(data['Gender'])
dummies
```

```python
data = pd.concat([data,dummies],axis=1)
data
```

```python
#lets drop the original gender column and "female" column from the created dummpy variables to remove multi-collinearlty
```

```python
data = data.drop(columns=['Gender','Female'])
```
